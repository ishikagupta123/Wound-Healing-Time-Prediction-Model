```python
import cv2
import numpy as np
import os


# Directory path containing the images
directory = '/content/drive/MyDrive/train_images'

# List all files in the directory
image_paths = [os.path.join(directory, filename) for filename in os.listdir(directory) if filename.endswith('.png')]



# Iterate over the image paths
for image_path in image_paths:
    # Load the wound image
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to segment the wound region
    _, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)

    # Calculate the average pixel intensity within the segmented wound region
    intensity = np.mean(gray[binary == 255])

     # Print the estimated intensity for each image
    print("Image Path:", image_path)
    print("Estimated Intensity:", intensity)
    print()
```

```python
import cv2
import numpy as np
import os

def find_coordinates(image, color):
    """Finds the coordinates of the specified color in the image.

    Args:
        image: The image to search.
        color: The color to search for.

    Returns:
        A list of the coordinates of the color in the image.
    """

    width, height, _ = image.shape
    coordinates = []
    for x in range(width):
        for y in range(height):
            if np.all(image[x, y] == color):
                coordinates.append((x, y))
    return coordinates

def process_images_in_directory(directory_path, color):
    """Processes all images in a directory and finds the coordinates of the specified color.

    Args:
        directory_path: The path to the directory containing the images.
        color: The color to search for.

    Returns:
        A dictionary where the keys are image filenames and the values are lists of coordinates of the color.
    """

    image_files = [f for f in os.listdir(directory_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    results = {}

    for image_file in image_files:
        image_path = os.path.join(directory_path, image_file)
        image = cv2.imread(image_path)
        coordinates = find_coordinates(image, color)
        results[image_file] = coordinates

    return results

if __name__ == "__main__":
    directory_path = "/content/drive/MyDrive/train_images"
    color = [255, 0, 0]
    image_coordinates_dict = process_images_in_directory(directory_path, color)

    for image_file, coordinates in image_coordinates_dict.items():
        print(f"Coordinates for {image_file}: {coordinates}")
```
